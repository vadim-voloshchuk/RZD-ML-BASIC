# Задача для компании "РЖД"

На железнодорожной станции необходимо создать программу для учета поездов, которые прибывают и отправляются. Программа должна обеспечивать следующие функции:

Создание и инициализация списка поездов:

1. Создать пустой список поезда.
Каждый поезд будет представлен в виде словаря со следующими данными:
- номер (уникальный номер поезда),
- направление (направление движения: "на восток" или "на запад"),
- время_прибытия (время прибытия поезда на станцию в формате "ЧЧ:ММ"),
- время_отправления (время отправления поезда со станции в формате "ЧЧ:ММ").
2. Добавление нового поезда:

Написать функцию добавить_поезд, которая принимает на вход параметры номер, направление, время_прибытия и время_отправления, и добавляет новый поезд в список поезда.
3. Вывод информации о поездах:

Написать функцию вывести_информацию_о_поездах, которая позволяет вывести список всех поездов на станции. Информация о каждом поезде должна быть отформатирована для наглядного представления.
4. Поиск поездов по направлению:

Написать функцию найти_поезда_по_направлению, которая принимает на вход направление ("на восток" или "на запад") и выводит список всех поездов, движущихся в указанном направлении.
5. Удаление поезда по номеру:

Написать функцию удалить_поезд_по_номеру, которая принимает на вход номер поезда и удаляет соответствующий поезд из списка поезда.
6. Обновление информации о поезде:

Написать функцию обновить_информацию_о_поезде, которая позволяет обновить информацию (например, время прибытия или отправления) о конкретном поезде на основе его номера.
7. Главное меню:

Создать главное меню, в котором пользователь может выбирать операции, выполняемые с помощью вышеуказанных функций. Программа должна предоставлять возможность добавления, просмотра, поиска, удаления и обновления информации о поездах.
8. Цикл выполнения:

Реализовать бесконечный цикл выполнения программы, в котором пользователь может выбирать операции и взаимодействовать с системой учета поездов.
